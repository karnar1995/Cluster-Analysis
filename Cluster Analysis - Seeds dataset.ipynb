## K Means

```python
from sklearn.cluster import KMeans
```

First various clusters are checked based on the SSE score

```python
# Run Kmeans for values of K from 2 to 10.
# Collect the SSE (using the inertia_ property) in an array for plotting later
SSE = []
for K in range(2,11):
    kmeans = KMeans(init='random', n_clusters=K)
    kmeans.fit(data1_scaled)
    SSE = SSE + [kmeans.inertia_]
    print(kmeans.inertia_)
    
```

Plotting the SSE measures against various clusters. (ELBOW Method)

```python
# Plot the SSE values we collected above
fig = plt.figure(figsize=(8, 4), dpi=100)
ax = fig.add_subplot(1, 1, 1)
ax.plot(list(range(2,11)), SSE, marker='.', markersize=10)
ax.set_title('SSE for different values of K')
plt.show()
```

The elbow from the above plot is not correctly seen but we are choosing the number of clusters as 3 as there is sharp decrease in SSE from 2 clusters to 3 clusters and then the slope decreases.

## Silhouette Score

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
```

```python
for K in range(2,10):
    
    # Cluster using Kmeans
    kmeans = KMeans(init='random', n_clusters=K)
    y = kmeans.fit_predict(data1_scaled)

    # Create side-by-side plots
    fig = plt.figure(figsize=(12, 4), dpi=100)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    # ------------------------------------------
    # The left subplot is the silhouette plot
    # ------------------------------------------

    ax1.set_xlim([-0.1, 1.0])
    vertical_spacing = (K+1)*10 # extra spacing to separate the silhouettes
    ax1.set_ylim([0, len(data1_scaled) + vertical_spacing])

    # The silhouette_score gives the average value for all the samples.
    silhouette_avg = silhouette_score(data1_scaled, y)
    print("For K = {}, the average silhouette score is {:.3f}".format(K, silhouette_avg))

    # Compute the silhouette scores for each instance/sample/object
    sample_silhouette_values = silhouette_samples(data1_scaled, y)

    y_lower = 10
    for i in range(K):
        # Aggregate and sort the silhouette scores for samples in cluster i
        ith_cluster_silhouette_values = sample_silhouette_values[y==i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        # color setup
        color = cm.nipy_spectral(float(i) / K)

        # Fill the area between two curves
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
            0, ith_cluster_silhouette_values,
            facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("Silhouette Plot")
    ax1.set_xlabel("Silhouette Score")
    ax1.set_ylabel("Cluster Label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # ------------------------------------------
    # The right plot shows the actual clusters formed
    # ------------------------------------------

    colors = cm.nipy_spectral(y.astype(float) / K)
    ax2.scatter(data1_scaled.iloc[:, 0], data1_scaled.iloc[:, 2], marker='.', s=70, lw=0, alpha=0.7, c=colors, edgecolor='k')

    # Mark the cluster centers
    centers = kmeans.cluster_centers_
    ax2.scatter(centers[:, 0], centers[:, 2], marker='+', c="black", alpha=1, s=200, edgecolor='k')

    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 2], marker='o', c="white", alpha=1, s=200, edgecolor='k')
    
    # Label the cluster centers
    for i, c in enumerate(centers):
        cl = '${}$'.format(i+1)
        ax2.scatter(c[0], c[2], marker=cl, alpha=1, s=50, c='k', edgecolor='')

    ax2.set_title("Clustered Data")
    ax2.set_xlabel("Area")
    ax2.set_ylabel("Asymmetry")

    plt.suptitle("Silhouette analysis for KMeans clustering with K = {}".format(K),
                 fontsize=14, fontweight='bold')

    plt.show()
```

From the Silhouette score we see that the number of clusters = 2 the score is high but the shape of the cones is not similar as one cone is very large. Hence we chose the number of clusters as 3 based on these factors.

The correlation score is computed below for number of clusters=3

```python
kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(data1_scaled)
```

```python
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.metrics.pairwise import pairwise_distances
from scipy.stats import pearsonr

ix = y_kmeans.argsort()


X_sorted = data1_scaled.iloc[ix]
y_sorted = y_kmeans[ix]

# Display unsorted distance matrix
dist_mat = squareform(pdist(X_sorted))
N = len(X_sorted)
plt.pcolormesh(dist_mat)
plt.colorbar()
plt.xlim([0,N])
plt.ylim([0,N])
plt.show()
```

```python
# Compute the pairwise distances among all points
dist_mx = pairwise_distances(data1_scaled)

# Function to convert distances to similarities in the range 0 to 1
def similarity(x):
    dmin = dist_mx.min()
    dmax = dist_mx.max()
    return(1-((x-dmin)/(dmax-dmin)))

# Compute the pairwise similiarities
sim_mx = similarity(dist_mx)

# Calculate incidence matrix
# 1 if point i and point j are in the same cluster, 0 otherwise
incd_mx = []
for i in y_kmeans:
    row = []
    for j in y_kmeans:
        if i == j:
            row.append(1)
        else:
            row.append(0)
    incd_mx.append(row)

incd_mx = np.array(incd_mx)

# Now we can compute the correlation between the incidence matrix
# and the similarity matrix:
corr, pvalue = pearsonr(np.array(sim_mx).reshape(-1,1), np.array(incd_mx).reshape(-1,1))
print(corr)
print(pvalue)
```

The correlation score is found to be 0.560

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from matplotlib.colors import ListedColormap
from sklearn.neighbors import NearestNeighbors
```

To find the hyperparameters eps and minpts we try calculating for various Nearest Neighbours

```python
# Create a NearestNeighbors object, fit the data, then get the 4 nearest neighbors

for k in range(3,10):
    nbrs = NearestNeighbors(n_neighbors=k).fit(data1_scaled)
    distances, indices = nbrs.kneighbors(data1_scaled)

    # Get the indexes of the array sorted by the 4th nearest neighbor
    sorted_ix = distances[:,k-1].argsort()

    # Extract the plot y values
    py = distances[sorted_ix,k-1]
    px = list(range(len(distances)))

    fig = plt.figure(figsize=(6, 6), dpi=100)
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(px, py) #, marker='.', s=25, edgecolor='k')
    ax.set_title('{} nearest neighbor distance'.format(k))
    ax.set_xlabel('Instances (sorted by Kth nearest neighbor distance)')
    ax.set_ylabel('Eps')
    ax.grid()
    
plt.show()
```

 From the above graphs we end up with an average value of eps=0.75 and min_samples=4

```python
for K in range(2,10):
    
    # Cluster using Kmeans
    kmeans = KMeans(init='random', n_clusters=K)
    y = kmeans.fit_predict(d_reduced)

    # Create side-by-side plots
    fig = plt.figure(figsize=(12, 4), dpi=100)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    # ------------------------------------------
    # The left subplot is the silhouette plot
    # ------------------------------------------

    ax1.set_xlim([-0.1, 1.0])
    vertical_spacing = (K+1)*10 # extra spacing to separate the silhouettes
    ax1.set_ylim([0, len(d_reduced) + vertical_spacing])

    # The silhouette_score gives the average value for all the samples.
    silhouette_avg = silhouette_score(d_reduced, y)
    print("For K = {}, the average silhouette score is {:.3f}".format(K, silhouette_avg))

    # Compute the silhouette scores for each instance/sample/object
    sample_silhouette_values = silhouette_samples(d_reduced, y)

    y_lower = 10
    for i in range(K):
        # Aggregate and sort the silhouette scores for samples in cluster i
        ith_cluster_silhouette_values = sample_silhouette_values[y==i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        # color setup
        color = cm.nipy_spectral(float(i) / K)

        # Fill the area between two curves
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
            0, ith_cluster_silhouette_values,
            facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("Silhouette Plot")
    ax1.set_xlabel("Silhouette Score")
    ax1.set_ylabel("Cluster Label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # ------------------------------------------
    # The right plot shows the actual clusters formed
    # ------------------------------------------

    colors = cm.nipy_spectral(y.astype(float) / K)
    ax2.scatter(d_reduced[:, 0], d_reduced[:, 1], marker='.', s=70, lw=0, alpha=0.7, c=colors, edgecolor='k')

    # Mark the cluster centers
    centers = kmeans.cluster_centers_
    ax2.scatter(centers[:, 0], centers[:, 1], marker='+', c="black", alpha=1, s=200, edgecolor='k')

    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o', c="white", alpha=1, s=200, edgecolor='k')
    
    # Label the cluster centers
    for i, c in enumerate(centers):
        cl = '${}$'.format(i+1)
        ax2.scatter(c[0], c[1], marker=cl, alpha=1, s=50, c='k', edgecolor='')

    ax2.set_title("Clustered Data")
    ax2.set_xlabel("Area")
    ax2.set_ylabel("Asymmetry")

    plt.suptitle("Silhouette analysis for KMeans clustering with K = {}".format(K),
                 fontsize=14, fontweight='bold')

    plt.show()
```

```python
# Compute the pairwise distances among all points
dist_mx = pairwise_distances(d_reduced)

# Function to convert distances to similarities in the range 0 to 1
def similarity(x):
    dmin = dist_mx.min()
    dmax = dist_mx.max()
    return(1-((x-dmin)/(dmax-dmin)))

# Compute the pairwise similiarities
sim_mx = similarity(dist_mx)

# Calculate incidence matrix
# 1 if point i and point j are in the same cluster, 0 otherwise
incd_mx = []
for i in y_kmeans:
    row = []
    for j in y_kmeans:
        if i == j:
            row.append(1)
        else:
            row.append(0)
    incd_mx.append(row)

incd_mx = np.array(incd_mx)

# Now we can compute the correlation between the incidence matrix
# and the similarity matrix:
corr, pvalue = pearsonr(np.array(sim_mx).reshape(-1,1), np.array(incd_mx).reshape(-1,1))
print(corr)
print(pvalue)
```
